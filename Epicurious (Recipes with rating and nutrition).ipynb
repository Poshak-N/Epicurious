```python
df=df.dropna()
```

```python
df.calories	=df.calories.fillna(df.calories.mean())
df.protein=df.protein.fillna(df.protein.mean())
df.fat=df.fat.fillna(df.fat.mean())
df.sodium=df.sodium.fillna(df.sodium.mean())
```

```python
df.shape
```

```python
df.head()
```

```python
df1=df[["rating","calories","protein","fat","sodium"]]
df1.head()
```

```python
for col in df1.columns:
    sns.boxplot(x=df1[col])
    plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
Q1=df1.quantile(0.25)
Q2=df1.quantile(0.75)
IQR=Q2-Q1

lower_bound=Q1-(1.5*IQR)
upper_bound=Q2+(1.5*IQR)

for col in df1.select_dtypes(include=np.number).columns:
    df1[col]=np.where((df1[col]<lower_bound[col]),lower_bound[col],df1[col])
    df1[col]=np.where((df1[col]>upper_bound[col]),upper_bound[col],df1[col])
```

```python
for col in df1.columns:
    sns.boxplot(x=df1[col])
    plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
corr=df1.corr()
corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
